#A)Look for the missing values in all the columns and either impute them (replace with mean,median, or mode) or drop them. Justify your action for this task.

```python
import pandas as pd
df = pd.read_csv("/content/train.csv")
df
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
missing_columns = ['Mileage', 'Engine', 'Power', 'Seats']

for col in missing_columns:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna(df[col].mode()[0])
    else:
        df[col] = df[col].fillna(df[col].mean())


```

```python
#removing the 'New_Price' because more than 80% of data is missing so i  dropping the column
df.drop('New_Price', axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
#dowloading updated missing values file
from google.colab import files
df.to_csv('cleaned_data_train_A.csv', index=False)
files.download('cleaned_data_train_A.csv')
```

#C)Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot encoded value.

```python
df_c=pd.read_csv("/content/cleaned_data_train_B.csv")
df_c
```

```python
print(df_c['Fuel_Type'].unique())
print(df_c['Transmission'].unique())
```

```python
# One-hot encode Fuel_Type (handles Diesel, Petrol, Electric)
fuel_dummies = pd.get_dummies(df_c['Fuel_Type'], prefix='Fuel')
df_c = pd.concat([df_c, fuel_dummies], axis=1)
df_c.drop('Fuel_Type', axis=1, inplace=True)
# Binary map for Transmission
df_c['Transmission'] = df_c['Transmission'].map({'Automatic': 0, 'Manual': 1})

# Convert all dummy columns to integers (0/1)
fuel_cols = ['Fuel_Diesel', 'Fuel_Petrol', 'Fuel_Electric']
df_c[fuel_cols] = df_c[fuel_cols].astype(int)

```

#E)Perform select, filter, rename, mutate, arrange and summarize with group by operations (or their equivalent operations in python) on this dataset.


```python
df_e = pd.read_csv("/content/cleaned_data_train_D.csv")
df_e
```

```python
df_selected = df_e[['Name', 'Year', 'Mileage', 'Fuel_Petrol', 'Transmission']]
df_filtered = df_e[df_e['Mileage'] > 20]
df_renamed = df_e.rename(columns={'Mileage': 'Mileage_kmpl', 'Transmission': 'Is_Manual'})
df_e['Car_Age'] = current_year - df_e['Year']
df_sorted = df_e.sort_values(by='Price', ascending=False)
df_summary = df_e.groupby(['Fuel_Petrol', 'Fuel_Diesel', 'Fuel_Electric'])[['Price', 'Mileage']].mean().reset_index()
# Complex summary statistics using agg() on one-hot encoded Fuel_Type columns
summary_stats_complex = df_e.groupby(['Fuel_Petrol', 'Fuel_Diesel', 'Fuel_Electric']).agg({'Price': ['mean', 'min', 'max']})


print("Selected Cols")
print(df_selected.head())

print("\nFiltered Data (Year > 2015):")
print(df_filtered.head())

print("\nRenamed Columns:")
print(df_renamed.head())

print("\nNew Column 'Car_Age':")
print(df_e[['Name', 'Year', 'Car_Age']].head())

print("\nSorted by Price:")
print(df_sorted.head())

print("\nSummary Statistics (Mean Price by Fuel Type):")
print(df_summary)

print("\nSummary Stats (Mean, Min, Max Price by Fuel Type):")
print(summary_stats_complex)


```
